DTSA 5502 Trees and Graphs: Basics

Instructor Sriram Sankaranarayanan

Course Description

This course will continue our exploration of algorithms and data-structures over trees and graphs. In particular, trees and graphs are used to represent relational data and networks such as computer networks, road networks, biological networks and so on. The course will focus on the following topics:

    Binary Search Tree Data Structure and Balanced Binary Search Trees. 

    Representing Graphs, Graph Traversals and their Applications.

    Spanning Trees over Graphs.

    Amortized Analysis and Disjoint Set Data Structure.

    Shortest Path Algorithms over Graphs.

The algorithmic techniques covered are quite fundamental and the student will learn the basic approaches to analyze these algorithms. They will also learn to translate their knowledge into developing python programs for applications.

Drops and Refunds

Each student has 14 days from a class start date or their enrollment date (whichever is later) to request a drop for 100% tuition refund. Students are only eligible to drop a course if they have not accessed the restricted content (timed proctored exam) or received a grade.  When a course is dropped under these conditions, it will not appear on the student’s record.
Withdrawal

Students who request to drop the course after the 14-day period and who have not accessed the timed proctored assessment may withdraw from the course but will not receive a refund. When a student withdraws from a course under these conditions, the student will receive a grade of W on their academic record. W grades have no bearing on the GPA and credit total.

Students who access a timed, proctored final exam are ineligible for a drop, withdrawal, or refund, and are assigned a grade.
