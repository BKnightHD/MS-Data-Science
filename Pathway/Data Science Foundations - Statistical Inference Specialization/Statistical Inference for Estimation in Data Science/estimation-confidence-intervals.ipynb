```r
# Given values
values <- c(2781, 2900, 3013, 2856, 2888)

# Calculate sample mean and sample standard deviation
mean_value <- mean(values)
std_dev <- sd(values)

# Calculate the confidence interval
n <- length(values)
z <- qnorm(0.975)  # 0.975 corresponds to (1 - 0.95) / 2 for a two-tailed test
margin_of_error <- z * (std_dev / sqrt(n))
lower_bound <- mean_value - margin_of_error
upper_bound <- mean_value + margin_of_error

# Output the result
cat("95% Confidence Interval: [", round(lower_bound, 2), ",", round(upper_bound, 2), "]")
```

## Confidence intervals for proportions and variances

```r
z <- pnorm(.975)
p = 112/250
n = 250
moe = z * sqrt(p * (1-p) / n)
round(p - moe , 4)
round(p + moe, 4)
```

```r
n1 = 150
n2 = 150
p1 = 78 / n1
p2 = 92 / n2
z = qnorm(.975)
truep = p2 - p1
v1 = p1 * (1-p1) / n1
v2 = p2 * (1-p2) / n2
moe = z * sqrt(v1 + v2)

round(truep - moe, 4)
round(truep + moe, 4)
```

```r
x = c(9.2229, 8.3665, 8.797058, 10.2195, 6.5629)
z = qnorm(.96)
n = sqrt(length(x))
xbar = mean(x)
s = sd(x)
moe = z * s / n
round(xbar - moe, 4)
round(xbar + moe, 4)
```

```r
# Given values
values <- c(9.2229, 8.3665, 8.797058, 10.2195, 6.5629)
alpha <- 0.08  # 92% confidence level

# Calculate sample variance
n <- length(values)
mean_value <- mean(values)
squared_diff <- (values - mean_value)^2
sample_variance <- sum(squared_diff) / (n - 1)

# Calculate chi-square values
chi_lower <- qchisq(alpha/2, df = n - 1)
chi_upper <- qchisq(1 - alpha/2, df = n - 1)

# Calculate confidence interval
lower_bound <- (n - 1) * sample_variance / chi_upper
upper_bound <- (n - 1) * sample_variance / chi_lower

# Output the result
cat("92% Confidence Interval for the true variance: (", round(lower_bound, 4), ",", round(upper_bound, 4), ")")
```
